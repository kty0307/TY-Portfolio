```python
# 환율 
USD_KRW = pd.read_csv('../알고리즘플젝/USD_KRW.csv')
df_1 = pd.DataFrame(USD_KRW)
df_1["date"] = pd.to_datetime(df_1["date"])  # 날짜 형식으로 변환
df_1 = df_1.set_index('date')
df_1["price"] = df_1["price"].str.replace(",", "").astype(float)  # 쉼표 제거 후 float로 데이터 타입 변환
df_u = df_1.resample('M').mean()  # 월별 평균 계산
df_u = df_u.rename(columns={"price": "USD_KRW"})
df_u
```

```python
# gold , kospi
df_g = pd.read_csv('Gold.csv')
df_k = pd.read_csv('kospi_cv.csv')
```

```python
df_g = df_g.rename(columns={"Date": "date", "Gold_close": "gold_close"})
df_g = pd.DataFrame(df_g)
df_g["date"] = pd.to_datetime(df_g["date"])  # 날짜 형식으로 변환
df_g = df_g.set_index('date')
df_g["gold_close"] = df_g["gold_close"].astype(float)  # 쉼표 제거 후 float로 데이터 타입 변환
df_g = df_g.resample('M').mean()  # 월별 평균 계산

print(df_g)
```

```python
# df_k
df_k = df_k.rename(columns={"날짜": "date", "kospi close":"kospi_close","kospi volume":"kospi_volume"})
df_k = pd.DataFrame(df_k)
df_k["date"] = pd.to_datetime(df_k["date"])  # 날짜 형식으로 변환
df_k = df_k.set_index('date')
df_k["kospi_close"] = df_k["kospi_close"].astype(float)
df_k["kospi_volume"] = df_k["kospi_volume"].astype(float)
df_k = df_k.resample('M').mean()  # 월별 평균 계산

print(df_k)
```

```python
df_u = df_u.reset_index()
df_g = df_g.reset_index()
df_k = df_k.reset_index()
merged_df = pd.merge(df_u, df_g, on="date").merge(df_k, on="date")
merged_df = merged_df.set_index('date')
merged_df.head()
```

```python
# SNP20.csv
df_s = pd.read_csv('SNP20.csv')
df_s = df_s.rename(columns={"close": "snp_close", "volume":"snp_volume"})
df_s = pd.DataFrame(df_s)
df_s["date"] = pd.to_datetime(df_s["date"])  # 날짜 형식으로 변환
df_s = df_s.set_index('date')
df_s["snp_close"] = df_s["snp_close"].astype(float)
df_s["snp_volume"] = df_s["snp_volume"].astype(float)
df_s = df_s.resample('M').mean()  # 월별 평균 계산
df_s = df_s.reset_index()
df_s
```

```python
merged_df = merged_df.reset_index()
merged_df = pd.merge(merged_df, df_s, on="date")
merged_df = merged_df.set_index('date')
merged_df
```

```python
bondncpi = pd.read_csv('../알고리즘플젝/bondncpi.csv')
bondncpi = bondncpi.rename(columns={"변환": "date"})
bondncpi = pd.DataFrame(bondncpi)
bondncpi["date"] = pd.to_datetime(bondncpi["date"])  # 날짜 형식으로 변환
bondncpi["kgb_1y_return"] = bondncpi["kgb_1y_return"].astype(float)
bondncpi["kgb_1y_volume"] = bondncpi["kgb_1y_volume"].str.replace(",", "").astype(float)
bondncpi["ktb_3m_return"] = bondncpi["ktb_3m_return"].astype(float)
bondncpi["ktb_3m_volume"] = bondncpi["ktb_3m_volume"].str.replace(",", "").astype(float)
bondncpi["cpi_index"] = bondncpi["cpi_index"].astype(float)
bondncpi = bondncpi.set_index('date')
bondncpi = bondncpi.resample('M').mean()
bondncpi = bondncpi.reset_index()
bondncpi
```
